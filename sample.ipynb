```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Load embedding model
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# --- 1. Base query ---

query = "Technology is changing business"

# --- 2. Expanded queries (LLM-generated or manually crafted) ---
expanded_queries = [
    query,
    "How is AI transforming companies",
    "Impact of digital innovation on business",
    "Modern tech trends affecting organizations",
    "Business changes driven by automation"
]

# --- 3. Corpus ---
documents = [
    "The quick brown fox jumps over the lazy dog.",
    "Artificial intelligence is transforming the world.",
    "I enjoy hiking in the mountains.",
    "The stock market saw significant gains today.",
    "Automation and AI are reshaping modern businesses."
]

# --- 4. Embed corpus once ---
doc_embeddings = model.encode(documents, convert_to_numpy=True)

# --- 5. Embed expanded queries ---
query_embeddings = model.encode(expanded_queries, convert_to_numpy=True)

# --- 6. Cosine similarity ---
def cosine_similarity(a, b):
    a_norm = a / np.linalg.norm(a, axis=1, keepdims=True)
    b_norm = b / np.linalg.norm(b, axis=1, keepdims=True)
    return np.dot(a_norm, b_norm.T)

# Compute similarity matrix: (num_queries x num_docs)
scores = cosine_similarity(query_embeddings, doc_embeddings)

# Aggregate scores across all expanded queries
final_scores = scores.max(axis=0)  # or mean(axis=0)

# --- 7. Rank results ---
top_k = 3
top_indices = np.argsort(final_scores)[::-1][:top_k]

print("\nBase Query:", query)
print("\nTop matches using Multi‑Query Expansion:")
for idx in top_indices:
    print(f"- Score: {final_scores[idx]:.4f} | {documents[idx]}")
```
